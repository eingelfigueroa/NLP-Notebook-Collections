```python
import os
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
import re
```

```python
os.chdir("C://Users/AikaS/Desktop")
text = open('rawData.txt', "r")
data = text.read()
```

```python
print(data)
```

```python
ensw = stopwords.words('english')
```

```python
print(ensw)
```

```python
textArray = word_tokenize(data.lower())
```

```python
filterArray = [item for item in textArray if item not in ensw]
```

```python
print(filterArray)
```

```python
stringFilter = ','.join(filterArray)
```

```python
print(stringFilter)
```

```python
words = set(nltk.corpus.words.words())
```

```python
stringString = re.sub('\W+','\n',stringFilter)

f = open ('newList.txt', "w")
f.write(stringString)
f.close()
```

```python
with open ('newList.txt') as fp:
    line = fp.readline()
    cnt = 1
    f = open('finalList.txt', "w")
    while line:
        data = line.strip()
        if data in words:
            print(data)
            f.write(data + "\n")
        line = fp.readline()
        cnt += 1
```

```python
finalText = open('finalList.txt', "r")
finalData = finalText.read()

tokenwords = word_tokenize(finalData)
countWords = Counter(tokenwords)
```

```python
print(countWords)
```

```python
countList = list(Counter((countWords)).items())
newList = '\n'.join([str(i) for i in countList])
```

```python
print(newList)
```
