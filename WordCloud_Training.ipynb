```python
import matplotlib.pyplot as plt
from wordcloud.wordcloud import WordCloud
```

```python
PText = open('positive.txt',"r")
positive = PText.read()

NText = open('negative.txt', "r")
negative = NText.read()
```

```python
import string
import nltk
### Run 3 times,
with open('clean_T10.txt', "r") as f:
    line = f.readline()
    fa = open('cleaned_T10.txt', "w")
    while line:
        data = line.strip()
        data = data.strip('[]')
        data = data.replace("'","")
        fa.write(data + "\n")
        line = f.readline()
```

```python
with open ('cleaned_T10.txt') as fp:
    line = fp.readline()
    cnt = 1
    f = open('stripped_T10v2.txt', "w")
    while line:
        for i,word in enumerate(line.split()):
            #print(word)
            f.write(word + "\n")
        line = fp.readline()
        cnt += 1
```

```python
text = open('stripped_T10v2.txt', "r")
data = text.read()
```

```python
plt.figure(figsize=(20,10))
wordcloud = WordCloud(background_color='white', mode="RGB", width=2000, height=1000).generate(data)
plt.title("Responses WordCloud")
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
with open ('stripped_T10v2.txt') as fp:
    line = fp.readline()
    cnt = 1
    f = open('positive_T10v2.txt', "w")
    while line:
        for i,word in enumerate(line.split()):
            if word in positive and word not in negative:
                #print(word)
                f.write(word + "\n")
        line = fp.readline()
        cnt += 1
```

```python
text = open('positive_T10v2.txt', "r")
data = text.read()

plt.figure(figsize=(20,10))
wordcloud = WordCloud(background_color='white', mode="RGB", width=2000, height=1000).generate(data)
plt.title("Responses WordCloud")
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
with open ('stripped_T10v2.txt') as fp:
    line = fp.readline()
    cnt = 1
    f = open('negative_T10v2.txt', "w")
    while line:
        for i,word in enumerate(line.split()):
            if word not in positive and word in negative:
                #print(word)
                f.write(word + "\n")
        line = fp.readline()
        cnt += 1
```

```python
text = open('negative_T10v2.txt', "r")
data = text.read()

plt.figure(figsize=(20,10))
wordcloud = WordCloud(background_color='white', mode="RGB", width=2000, height=1000).generate(data)
plt.title("Responses WordCloud")
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
