```python

import os
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
import re
```

```python
text = open('T2.txt', "r")
data = text.read()
data
```

```python
ensw = stopwords.words('english')
```

```python
textArray = word_tokenize(data.lower())
filterArray = [item for item in textArray if item not in ensw]
filterArray = [i for i in filterArray if len(i) > 2]
stringFilter = ','.join(filterArray)
words = set(nltk.corpus.words.words())
```

```python
stringString = re.sub('\W+','\n',stringFilter)

f = open ('newT2.txt', "w")
f.write(stringString)
f.close()
```

```python
with open ('newT2.txt') as fp:
    line = fp.readline()
    cnt = 1
    f = open('finalT2.txt', "w")
    while line:
        data = line.strip()
        if data in words:
            print(data)
            f.write(data + "\n")
        line = fp.readline()
        cnt += 1
```

```python
PText = open('positive.txt',"r")
positive = PText.read()

NText = open('negative.txt', "r")
negative = NText.read()
```

```python
with open ('finalT2.txt') as fp:
    line = fp.readline()
    cnt = 1
    f = open('positiveT2.txt', "w")
    while line:
        data = line.strip()
        if data in positive:
            print(data)
            f.write(data + "\n")
        line = fp.readline()
        cnt += 1
```

```python
PLText = open('positiveT2.txt', "r")
positiveList = PLText.read()

textArray = word_tokenize(positiveList.lower())
filterArray = [item for item in textArray if item not in negative]


stringFilter ='\n'.join(filterArray)
stringFilter = Counter(filterArray)

print(stringFilter)
```

```python
tokenwords = word_tokenize(positiveList)
countList = list(Counter((tokenwords)).items())
newPositiveList = '\n'.join([str(i) for i in countList])
```

```python
NLText = open('negativeT1.txt', "r")
NegativeList = NLText.read()

textArray = word_tokenize(NegativeList.lower())
filterArray = [item for item in textArray if item not in positive]


stringFilter ='\n'.join(filterArray)
stringFilter = Counter(filterArray)

print(stringFilter)
```
