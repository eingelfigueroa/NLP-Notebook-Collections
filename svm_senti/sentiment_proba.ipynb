```python
from nltk.tokenize import TweetTokenizer

def tokenize(text): 
    tknzr = TweetTokenizer()
    return tknzr.tokenize(text)
```

```python
import pickle
model = pickle.load(open('svm_model.pkl','rb'))
test = ["im very satisfied with the service"]
```

```python
import numpy as np

def tanh(x, derivative=False):
    if (derivative == True):
        return (1 - (x ** 2))
    return np.tanh(x)
```

```python
test = model.predict_proba(test)

flat = tanh(test)
print(test)
flat = flat.tolist()

for i in flat:
    if i[0] > i[1]:
        num = tanh(i[0]) 
    else:
        num = tanh(-i[1])
    

    print(num)
    
```
