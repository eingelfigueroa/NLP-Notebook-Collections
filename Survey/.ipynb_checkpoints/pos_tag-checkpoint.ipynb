```python
import pandas as pd
import numpy as np
import nltk
nltk.download('vader_lexicon')
nltk.download('stopwords')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
```

```python
df = pd.read_csv('opinion_survey.csv')
df = df.replace(np.nan, 'Neutral', regex=True) ### Some values are missing, use "Neutral" for missing values
df = df.drop(df.columns[[0]],axis=1) ### Drop extra index
test_df = df.iloc[0:, 0:] # Just for renaming purposes so that df variable can be used for testing
test_df.head()
```

```python
from nltk.corpus import stopwords
stoplist = stopwords.words('english') + ['though']
```

```python
def getNgrams(wordlist, n):
    ngrams = []
    for i in range(len(wordlist)-(n-1)):
        ngrams.append(wordlist[i:i+n])
    
    new_words = list(filter(lambda w: w not in stoplist, ngrams))
    
    return ngrams
```

```python
test = 'Im very reluctant but i think it can be done'
getNgrams(test.split(),3)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
c_vec = CountVectorizer(stop_words=stoplist, ngram_range=(3,6))
comp = []
col_range = len(test_df.columns) # number of columns

for i in range(0,1):
    col = test_df.columns[i] # The current column
    ngrams = c_vec.fit_transform(test_df[col])
    vocab = c_vec.vocabulary_
    test_df['scores'] = test_df[col].apply(lambda x: sid.polarity_scores(x)) ## Get polarity score of every Column
    compound = test_df['scores'].apply(lambda score_dict: score_dict['compound']) ## Extract the compound from the results
    test_df = test_df.drop('scores', 1) # Drop score DF in every iteration
    compound = sum(compound)/140 # Get the mean compound of each columns
    comp.append(compound) # Save mean and append to list
    
df_ngram = pd.DataFrame(vocab.items()).rename(columns={0:'ngram'})
df_ngram.loc[0:]
```

```python
comp = []

df_ngram['sentiment'] = df_ngram['ngram'].apply(lambda x: sid.polarity_scores(x))
compound = df_ngram['sentiment'].apply(lambda score_dict: score_dict['compound'])
#print(test_df['sentiment'][139])
#print(df_ngram['ngram'][139])
df_ngram = df_ngram.drop('sentiment', 1) # Drop score DF in every iteration
compound = sum(compound)/len(compound) # Get the mean compound of each columns
comp.append(compound) # Save mean and append to list
print("Sentiment Score: ",comp)
df_ngram
```
