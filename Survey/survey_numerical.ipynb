```python
import pandas as pd
import numpy as np
df = pd.read_csv('likert_survey.csv')
df = df.drop(df.columns[[0,1]], axis=1)
#df = df.loc[1:]
df.shape 
```

```python
df.head()
```

### Check if there is Null Values

```python
check_nan_in_df = df.isnull().values.any()
check_nan_in_df
```

### Aggregated Likert Results

```python
Agg_results = pd.get_dummies(df.stack()).groupby(level=1).sum().reset_index() ### Distorts Original Positioning of columns and title, cannot find in documentation why it does this
#df = counts[counts.columns.drop(list(counts.filter(regex="[0-9]")))]
Agg_results.head()
```

### Sort columns by original position

```python
sorter = df.columns.tolist() ### List of correctly positioned titles
sorterIndex = dict(zip(sorter, range(len(sorter)))) ## Dictionary of titles with corresponding position number


Agg_results['ID'] = Agg_results['index'].map(sorterIndex) ### Create ID Column where its number will be based on 
                                                          ### the position number of the original positioning
                                                          ### of the titles
Agg_results = Agg_results.sort_values('ID',ascending = True) ## Sort DF by ID Column
Agg_results = Agg_results.drop('ID',axis=1) ## Drop ID column
Agg_results = Agg_results.reset_index() ## Reset index due to the original index numbers will be scrambled
Agg_results= Agg_results.drop(Agg_results.columns[[0]], axis=1) ## Drop old index Number
Agg_results.head()
```

### Columns names was arranged alphabetically, reorder to original order

```python
order = ['index','Strongly Agree', 'Agree', 'Disagree','Strongly Disagree'] ## Original Order
Agg_results = Agg_results.reindex(columns=order) ## Reindexing
Agg_results.reset_index(drop=True)
```

### Convert DF to List, but iteration of DF usually uses iterrows, but iterrows has a high Big 0 complexity

```python
likert_list = Agg_results.values.tolist()
likert_results = [i[1:] for i in likert_list] ## List comprehension of likert results = [[SA,A,D,SD],...]
likert_results
```

```python
likert_values = np.array([4,3,2,1]) ## Likert numerical Values
summed = likert_results * likert_values # Multiply values to number of results

rows = len(summed);  
cols = len(summed[0]);  

TOTAL_COLS = 140

#Create list to insert in DF later
numerical_value = []

#Calculates sum of each row of given matrix  
for i in range(0, rows):  
    sumRow = 0;  
    for j in range(0, cols):  
        sumRow = sumRow + summed[i][j];  
    print("Sum of " + str(i+1) +" row: " + format(sumRow/TOTAL_COLS, '.2f'))
    numerical_value.append(format(sumRow/TOTAL_COLS, '.2f'))
```

```python
## Insert List to DF
Agg_results["Numerical Value"] = numerical_value
Agg_results
```

```python
Agg_results.to_csv('survey_numerical_value.csv')
```
