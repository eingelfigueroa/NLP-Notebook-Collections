```python
import pandas as pd
import numpy as np
import scipy.stats
```

```python
df = pd.read_csv('opinion_sentiment.csv')
corr = df.drop(df.columns[[1,2,3,4]],axis=1)
corr
```

```python
#Aggregate Same Likert and Opinions first
# Manually fix inequal rows
'''
Numerical_mean  = corr.iloc[0:, 1].mean()
Sentiment_mean  = corr.iloc[0:, 2].mean()
means = [Numerical_mean,Sentiment_mean]
corr.loc[21] = ['Average'] + means
'''
                        ###Row  #Col
numerical_mean_first  = corr.iloc[13:17, 1].mean()
sentiment_mean_first = corr.iloc[13,2]

numerical_mean_sec  = corr.iloc[20:23, 1].mean()
sentiment_mean_sec = corr.iloc[20,2]

numerical_mean_third  = corr.iloc[24:27, 1].mean()
sentiment_mean_third = corr.iloc[24,2]

means1 = [numerical_mean_first,sentiment_mean_first]
corr.loc[29] = ['1st aggregation'] + means1

means2 = [numerical_mean_sec,sentiment_mean_sec]
corr.loc[30] = ['2nd aggregation'] + means2

means3 = [numerical_mean_third,sentiment_mean_third]
corr.loc[31] = ['3rd aggregation'] + means3

Numerical_mean  = corr.iloc[0:, 1].mean()
Sentiment_mean  = corr.iloc[0:, 2].mean()
means = [Numerical_mean,Sentiment_mean]
corr.loc[32] = ['Average'] + means

fixed_df = corr.drop([13,14,15,16,20,21,22,24,25,26])

fixed_df.reset_index(drop=True)
fixed_df.to_csv("fixed_numerical.csv", index=False)
fixed_df
```

```python
x_var = fixed_df.iloc[0:-1, 1]
y_var = fixed_df.iloc[0:-1, 2]
y_var
```

```python
pearson_r = scipy.stats.pearsonr(x_var, y_var) #Pearson Correlation Analysis
pearson_r = np.reshape(pearson_r, (1, 2))
corr_analysis = pd.DataFrame(pearson_r, columns=["R", "p-Value"])
corr_analysis
```
