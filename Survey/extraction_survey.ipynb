## Load merged surveys and remove irrelevant information(indices, timestamp, last three questions)

```python
new_df = pd.read_csv('Summary_survey.csv')
cropped_df = new_df.iloc[0:, 3:-3]
cropped_df.to_csv('cropped_survey.csv')
cropped_df
```

### Remove redundance question
### Extract Likert questions by dropping questions asking for opinions
### Some opinions has 2 or more likert questions between them, so stop and manually skip over these

```python
import numpy as np
cropped_df = new_df.iloc[0:, 3:31:2] ## start at col 3, stop at col 31 and skip every second column
dropped = cropped_df.drop(cropped_df.columns[[4,5]], axis=1) # Redundant Question
dropped['row_num'] = dropped.reset_index().index # Make Row Num

first_column = dropped.pop('row_num') 
dropped.insert(0, 'ID', first_column)
first_df = dropped
first_df
```

### At every stop, continue to skip over multiple likert questions and remove the opinion
### Create a DF in every stop and generate an ID num for merging later

```python
def make_num(df):
    df['row_num'] = df.reset_index().index # Make Row Num
    first_column = df.pop('row_num') 
    df.insert(0, 'ID', first_column)
    
    return df

second_df = new_df.iloc[0:, 31:35]
new = make_num(second_df)
new
```

### Continue manual process until done

```python
second_df = new_df.iloc[0:, 31:35]
second_df = make_num(second_df)

third_df = new_df.iloc[0:, 36:42:2]
third_df = make_num(third_df)

fourth_df = new_df.iloc[0:, 42:45]
fourth_df = make_num(fourth_df)

fifth = new_df.iloc[0:, 46:-3]
fifth = fifth.drop(fifth.columns[[1,5,7]], axis=1)
fifth = make_num(fifth)
```

### Merge trimmed dataframes

```python
from functools import reduce

df_test = [first_df,second_df,third_df,fourth_df,fifth]

df_final = reduce(lambda one,two: pd.merge(one,two, on='ID'), df_test)
likert_df = df_final.drop(df_final.columns[[3]], axis=1)
```

### Check if the titles are complete

```python
likert_title = likert_df.columns[1:].tolist()
for i in range(0,len(likert_title)):
    print(i+1, likert_title[i])
```

### Counter Check with original DF

```python
cropped_df = new_df.iloc[0:, 3:-3] #Remove Irrelevant Titles
opinion_title = cropped_df.drop(cropped_df.columns[[4,5]], axis=1) #Remove Redundant Titles

new_title = opinion_title.drop(likert_title, axis=1) # Remove Likert Titles
new_title.head()
```
