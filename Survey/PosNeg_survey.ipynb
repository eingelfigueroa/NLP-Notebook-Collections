```python
opinion = pd.read_csv('opinion_survey.csv')
df = opinion.drop(opinion.columns[[0]],axis=1)
print(opinion.shape)
df.tail()
```

```python
df = df.replace(np.nan, 'Neutral', regex=True)
```

```python
#DF to List
test = []
cols = len(df.columns)

for i in range(cols): 
    test_1 = []
    for j in range(len(df)):
        opi = df.iloc[j, i]
        #print(opi)
        test_1.append(opi)
    test.append(test_1)
```

```python
# Create Text file to store opinions
with open('opinion.txt','w') as file:
    for i in test:
        for j in i:
            file.write('%s\n' % j)
```

```python
import os
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
import re

nltk.download('stopwords')
nltk.download('words')
nltk.download('punkt')
```

```python
# Open PosNeg File

PText = open('positive.txt',"r")
positive = PText.read()

NText = open('negative.txt', "r")
negative = NText.read()
```

```python
# Remove stopwords and Filter Words

text = open('opinion.txt', "r")
data = text.read()

ensw = stopwords.words('english')

textArray = word_tokenize(data.lower())
filterArray = [item for item in textArray if item not in ensw]
filterArray = [i for i in filterArray if len(i) > 3]
stringFilter = ','.join(filterArray)
words = set(nltk.corpus.words.words())


stringString = re.sub('\W+','\n',stringFilter)

f = open ('new_opinion.txt', "w")
f.write(stringString)
f.close()
```

```python
# Filter if word is really a word
with open ('new_opinion.txt') as fp:
    line = fp.readline()
    cnt = 1
    f = open('final_opinion.txt', "w")
    while line:
        data = line.strip()
        if data in words:
            #print(data)
            f.write(data + "\n")
        line = fp.readline()
        cnt += 1
```

```python
# Create positive words

with open ('final_opinion.txt') as fp:
    line = fp.readline()
    cnt = 1
    f = open('positive_opinion.txt', "w")
    while line:
        data = line.strip()
        if data not in negative and data in positive:
            #print(data)
            f.write(data + "\n")
        line = fp.readline()
        cnt += 1
```

```python
# Tokenize Positive Words

PLText = open('positive_opinion.txt', "r")
positiveList = PLText.read()

textArray = word_tokenize(positiveList.lower())
filterArray = [item for item in textArray if item not in negative]


stringFilter ='\n'.join(filterArray)
stringFilter = Counter(filterArray)

print(stringFilter)
```

```python
# List Tokenized Pos Word

tokenwords = word_tokenize(positiveList)
countList = list(Counter((tokenwords)).items())
newPositiveList = '\n'.join([str(i) for i in countList])
print(newPositiveList)
```

```python
## manually Remove slippery positive words
post = ["easily","easy","agree"]

with open ('final_opinion.txt') as fp:
    line = fp.readline()
    cnt = 1
    f = open('negative_opinion.txt', "w")
    while line:
        data = line.strip()
        if data in negative and data not in post:
            #print(data)
            f.write(data + "\n")
        line = fp.readline()
        cnt += 1
```

```python
# Tokenize Negative Words

NLText = open('negative_opinion.txt', "r")
NegativeList = NLText.read()

textArray = word_tokenize(NegativeList.lower())
filterArray = [item for item in textArray if item not in positive]


stringFilter ='\n'.join(filterArray)
stringFilter = Counter(filterArray)

print(stringFilter)
```

```python
tokenwords = word_tokenize(NegativeList)
countList = list(Counter((tokenwords)).items())
newNegativeList = '\n'.join([str(i) for i in countList])
print(newNegativeList)
```

```python
finalNegative = re.sub("[)!@''#$(]", "", newNegativeList)
finalPositve = re.sub("[)!@''#$(]", "", newPositiveList)
```

```python
p = open('positiveData_opinion.csv', "w")
n = open('negativeData_opinion.csv', "w")

p.write(finalPositve)
n.write(finalNegative)

p.close()
n.close()
```
